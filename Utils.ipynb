```python
import cv2
import csv

import matplotlib.pyplot as plt
```

```python
def read_img(img_name, img_size):
    image = cv2.imread('train_/' + img_name)
    image = cv2.resize(image, (img_size, img_size))
    plt.imshow(image)
    return image        
```

```python
img_test = read_img('scan_0000.png', 256)
```

```python
def label_data():
    with open('csv/train.csv', 'r') as fobj:
        reader = csv.reader(fobj)
        next(reader, None)
        j = 0
        for i in reader:
            label = int((i[5].split('_'))[1])
            img_name = i[4]
            image = read_img(img_name, 256)
            
            # save to labelled directory
            cv2.imwrite('train_data/'+str(label-1)+'/'+img_name, image)
            print('train_data/'+str(label-1)+'/'+img_name)
```

```python
label_data()
```
